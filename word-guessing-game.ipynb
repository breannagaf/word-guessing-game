```python
# Word Guessing Game Simulation 
# This notebook contains the complete, interactive command-line Word-Guessing Game. The program selects a random word from a hidden list, and the player must guess the word letter-by-letter before running out of lives.
```

```python
import random
```

```python
## Step 1: Initializing Game Parameters and Word Bank
#Below, we define our word bank, set the maximum number of lives, and create a helper function to select a random word.
```

```python
def get_secret_word():
    """Selects and returns a random word from a list."""
    word_bank = ["strawberry", "jupyter", "chemistry", "blue", "crochet", "chocolate"]
    return random.choice(word_bank)
```

```python
## Step 2: Creating the Game Display State
# To hide the secret word, we represent un-guessed letters as underscores `_`. As the player guesses correctly, those underscores are updated with the actual letters.
```

```python
def display_current_state(secret_word, guessed_letters):
    """
    Creates a visual representation of the secret word, 
    revealing only the letters that have been guessed.
    """
    display_list = []
    for letter in secret_word:
        if letter in guessed_letters:
            display_list.append(letter)
        else:
            display_list.append("_")
    return " ".join(display_list)
```

```python
## Step 3: Core Game Execution Loop
# Here is the main control loop. Handle user inputs, check if they are valid alphabetic characters, update lives and guessed lists, and determine win/loss conditions.
```

```python
def play_game():
    # 1. Choose the secret word and initialize variables
    secret_word = get_secret_word()
    guessed_letters = set()
    lives_remaining = 6
    
    print("Welcome to the Word-Guessing Game!")
    print("---------------------------------")
    
    # 2. Start the interactive loop
    while lives_remaining > 0:
        # Display current progress
        current_display = display_current_state(secret_word, guessed_letters)
        print(f"\nWord: {current_display}")
        print(f"Guessed letters: {', '.join(sorted(guessed_letters)) if guessed_letters else 'None'}")
        print(f"Lives remaining: {lives_remaining}")
        
        # Win Condition Check
        if "_" not in current_display:
            print(f"\n🎉 Congratulations! You guessed the word: '{secret_word}'!")
            return
            
        # 3. Get and validate user input
        user_guess = input("Guess a letter: ").strip().lower()
        
        # Validation checks
        if len(user_guess) != 1 or not user_guess.isalpha():
            print("❌ Invalid input. Please enter exactly one single letter.")
            continue
            
        if user_guess in guessed_letters:
            print("⚠️ You already guessed that letter! Try another one.")
            continue
            
        # 4. Process the guess
        guessed_letters.add(user_guess)
        
        if user_guess in secret_word:
            print(f"✅ Great guess! '{user_guess}' is in the word.")
        else:
            lives_remaining -= 1
            print(f"❌ Sorry, '{user_guess}' is not in the word.")
            
    # Loss Condition
    print(f"\n💀 Game over! You ran out of lives. The word was '{secret_word}'.")
```

```python
## Step 4: Run the Interactive Game
# Execute the cell below to start playing!
```

```python
play_game()
```
